CREATE A CATEGORICAL VARIABLE FOR THE HEART EJECTION FRACTION

```python
import numpy as np
mortality_risk["EF_Cat"] = pd.cut(mortality_risk["EF"],
                    bins=[0, 29, 39, 52, np.inf],
                    labels=["Severe","Moderate","Mild","Normal"])
```

```python
mortality_risk["EF_Cat"].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(mortality_risk, mortality_risk["EF_Cat"]):
    strat_train_set = mortality_risk.loc[train_index]
    strat_test_set = mortality_risk.loc[test_index]
```

```python
strat_test_set["EF_Cat"].value_counts() / len(strat_test_set)
```

```python
# now that the data is split is stratified, you can make the data sets back to 'normal'

for set_ in (strat_train_set, strat_test_set):
    set_.drop("EF_Cat", axis = 1 , inplace = True)
```

Building A Machine learning algorithms to predict EF 

```python
#variables of interests
var = strat_train_set.drop("EF",axis=1)
#Data with the Highest corrolation with EF - OPTIONAL
var = strat_train_set[["NT-proBNP","Anion gap","Systolic blood pressure","gendera"]]
#predictors
pred = strat_train_set["EF"].copy()
```

Data Cleaning - Step already done with the selection of the non-null data categories and with the highest correlation with ejection fraction


Transformation Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler())
    ])

var_prepared = num_pipeline.fit_transform(var)
```

Training an Evaluationg on the training set

```python
## LINEAR REGRESSION ##
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(var_prepared,pred)

## TEST WITH THE TRAIN SET ##
x_test_one = var.iloc[:5]
y_test_one = pred.iloc[:5]

linRegModel_test_1 = num_pipeline.transform(x_test_one)
print("Predictions: ", lin_reg.predict(linRegModel_test_1))
print("Real EF: ", list(y_test_one))
```

```python
## TEST ON ON THE WHOLE DATA SET AND MEASURE THE ERROR WITH RMSE ##
from sklearn.metrics import mean_squared_error

EF_prediction = lin_reg.predict(var_prepared)
lin_mse = mean_squared_error(pred, EF_prediction)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
## A difference of 12% is not good enough ##
## A more powerfull model will be tested ##

## DECISION TREE REGRESSOR ##
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(var_prepared,pred)

## Now that the model is train, time to predict ##
EF_prediction = tree_reg.predict(var_prepared)
tree_mse = mean_squared_error(pred, EF_prediction)
tree_rmse = np.sqrt(tree_mse)
tree_rmse

```

CROSS VALIDATION with k-fold cross-validation and DECISION TREE REGRESSOR

```python
## k-fold cross-validation ##
from sklearn.model_selection import cross_val_score

tree_scores = cross_val_score(tree_reg, var_prepared, pred,scoring="neg_mean_squared_error",cv=10)
tree_rmse_scores = np.sqrt(-tree_scores)

## Let's look at results ##
def display_scores(scores):
    print("Scores: ", scores)
    print("Mean: ", scores.mean())
    print("Standart deviation", round(scores.std(),2))
display_scores(tree_rmse_scores)
```

COMPARE PERFORMANCES OF DECISION TREE AND LINEAR REGRESSION

```python
lin_scores = cross_val_score(lin_reg, var_prepared,pred,scoring="neg_mean_squared_error",cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)

display_scores(lin_rmse_scores)
```

CROSS VALIDATION with k-fold cross-validation and RANDOM FOREST REGRESSOR

```python
## RANDOM FOREST REGRESSOR ##
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(var_prepared,pred)

## Now that the model is train, time to predict ##
EF_prediction = forest_reg.predict(var_prepared)
forest_mse = mean_squared_error(pred, EF_prediction)
forest_rmse = np.sqrt(forest_mse)
forest_rmse # -> 4.777701326801313

## k-fold cross-validation ##


forest_scores = cross_val_score(forest_reg, var_prepared, pred,scoring="neg_mean_squared_error",cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```

FINE-TUNE : Grid search

```python
from sklearn.model_selection import GridSearchCV
##note -> n_estimatorsint : The number of trees in the forest || max_features : The number of features to consider when looking for the best split:

param_dist = [
    {'n_estimators':[1,10,100,1000],'max_features': [1,2,3,4],'bootstrap':[True]}
]

grid_search = GridSearchCV(forest_reg,param_dist,cv=5,scoring='neg_mean_squared_error',return_train_score=True)
grid_search.fit(var_prepared,pred)

## Let's see the resutls
cvres = grid_search.cv_results_
for mean_scores, prarams, in zip(cvres["mean_test_score"],cvres["params"]):
    print(np.sqrt(-mean_scores),prarams)
```

EVALUATE THE IMPORTANCE OF EACH ATTRIBUTE

```python
features_importance = grid_search.best_estimator_.feature_importances_
a_ = ['NT-proBNP', 'Anion gap', 'Systolic blood pressure', 'gendera']
sorted(zip(features_importance,a_),reverse=True)
```

EVALUATE ON TEST SET

```python
final_model = grid_search.best_estimator_

var_test = strat_test_set.drop('EF', axis=1)
#Data with the Highest corrolation with EF - OPTIONAL
var_test = strat_test_set[["NT-proBNP","Anion gap","Systolic blood pressure","gendera"]]
pred_test = strat_test_set['EF'].copy()

var_test_prepared = num_pipeline.fit_transform(var_test)

final_prediction = final_model.predict(var_test_prepared)
final_mse = mean_squared_error(pred_test,final_prediction)
final_rmse = np.sqrt(final_mse)# -> 12.504368614895538 
final_rmse
```

```python
## CONFIDENCE INTERVALS ##
from scipy import stats
confidence = 0.95
squared_errors = (final_prediction - pred_test)**2
np.sqrt(stats.t.interval(confidence, len(squared_errors)-1,
                        loc=squared_errors.mean(),
                         scale=stats.sem(squared_errors)))
#->[11.07250765, 13.78832996]
```
